Gráficas de Radio vs Tiempo y Ángulo vs Tiempo:

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

#Vector de tiempo:
t = np.linspace(0,8,100)

#Constantes de la ecuación de movimiento:
r0 = 0.5
w = 1

#Funciones paramétricas dependientes del tiempo para el radio y el ángulo:
def r(t):
    return r0/2 * np.e**(w*t) + r0/2 * np.e**(-w*t)

def phi(t):
    return w*t


#Gráficas:
plt.figure(1, dpi=100)
plt.plot(t,r(t), color="black")
plt.grid()
plt.title("Radio vs Tiempo")
plt.xlabel("Tiempo (t)")
plt.ylabel("Radio r(t)")
plt.show()

plt.figure(2, dpi=100)
plt.plot(t,phi(t), color="black")
plt.grid()
plt.title("Ángulo vs Tiempo")
plt.xlabel("Tiempo (t)")
plt.ylabel("Ángulo φ(t)")
plt.show()
```

Gráfica Radio vs Ángulo en el intérvalo $[0,2\pi )$:

```python
#Vector de ángulo:
v_phi = np.linspace(0,2*np.pi,100)

#Función del radio dependiente del ángulo
def r_phi(v_phi):
    return r0/2 * np.e**(v_phi) + r0/2 * np.e**(-v_phi)

#Gráfica:
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(v_phi, r_phi(v_phi), color="black")
fig.dpi = 100
plt.title("Radio vs Ángulo en el intervalo [0,2π)")
plt.show()

```

Gráfica curva paramétrica:

```python
#CONSTANTES:
r0 = 0.000001
w = 1

#PARÁMETROS GRAFICA 1:
t1 = np.linspace(0,10,100)
x_1= r(t1)*np.cos(phi(t1))
y_1 = r(t1)*np.sin(phi(t1))


#GRÁFICA CORTA:
plt.figure(3, dpi=100)
plt.plot(x_1,y_1,color="black")
plt.title("Trayectoria para un tiempo t=10")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()


#CONSTANTES GRÁFICA 2:

r0 = 0.000001
w = 0.1

#PARÁMETROS GRÁFICA 2:

t2 = np.linspace(0,1000,10000)
x_2= r(t2)*np.cos(phi(t2))
y_2 = r(t2)*np.sin(phi(t2))

plt.figure(4, dpi=100)
plt.plot(x_2,y_2, color="black")
plt.xscale("symlog")
plt.yscale("symlog")
plt.title("Trayectoria para un tiempo grande en escala logarítmica")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

Variaciones Omega:

```python
tw = np.linspace(0,1000,10000)

w = 0.05
x_w1= r(tw)*np.cos(phi(tw))
y_w1 = r(tw)*np.sin(phi(tw))
w = 0.07
x_w2= r(tw)*np.cos(phi(tw))
y_w2 = r(tw)*np.sin(phi(tw))
w = 0.09
x_w3= r(tw)*np.cos(phi(tw))
y_w3 = r(tw)*np.sin(phi(tw))
w = 0.1
x_w4= r(tw)*np.cos(phi(tw))
y_w4 = r(tw)*np.sin(phi(tw))

fig5, axs = plt.subplots(2, 2)
fig5.suptitle("Trayectoria logarítimica para distintos valores de ω")
axs[0,0].plot(x_w1,y_w1, color="black")
axs[0,0].set_title("ω = 0.1")
axs[0,0].set_xscale("symlog")
axs[0,0].set_yscale("symlog")
axs[0,1].plot(x_w2,y_w2, color="black")
axs[0,1].set_title("ω = 0.2")
axs[0,1].set_xscale("symlog")
axs[0,1].set_yscale("symlog")
axs[1,0].plot(x_w3,y_w3, color="black")
axs[1,0].set_title("ω = 0.3")
axs[1,0].set_xscale("symlog")
axs[1,0].set_yscale("symlog")
axs[1,1].plot(x_w4,y_w4, color="black")
axs[1,1].set_title("ω = 0.4")
axs[1,1].set_xscale("symlog")
axs[1,1].set_yscale("symlog")

for ax in axs.flat:
    ax.label_outer()
    ax.label_inner()

plt.show()


```
